```python
jan20 = trip_jan20.sample(n = 1000, random_state = 2)
feb20 = trip_feb20.sample(n = 1000, random_state = 2)
mar20 = trip_mar20.sample(n = 1000, random_state = 2)
apr20 = trip_apr20.sample(n = 1000, random_state = 2)
may20 = trip_may20.sample(n = 1000, random_state = 2)
jun20 = trip_jun20.sample(n = 1000, random_state = 2)
jul20 = trip_jul20.sample(n = 1000, random_state = 2)
aug20 = trip_aug20.sample(n = 1000, random_state = 2)
sep20 = trip_sep20.sample(n = 1000, random_state = 2)
oct20 = trip_oct20.sample(n = 1000, random_state = 2)
nov20 = trip_nov20.sample(n = 1000, random_state = 2)
dec20 = trip_dec20.sample(n = 1000, random_state = 2)
```

```python
def mean_hour(month):
    time = sum(month)/len(month)
    seconds = int(time)
    minutes = seconds/60
    return minutes
```

```python
jan19_avg_dur = mean_hour(jan19['tripduration'])
feb19_avg_dur = mean_hour(feb19['tripduration'])
mar19_avg_dur = mean_hour(mar19['tripduration'])
apr19_avg_dur = mean_hour(apr19['tripduration'])
may19_avg_dur = mean_hour(may19['tripduration'])
jun19_avg_dur = mean_hour(jun19['tripduration'])
jul19_avg_dur = mean_hour(jul19['tripduration'])
aug19_avg_dur = mean_hour(aug19['tripduration'])
sep19_avg_dur = mean_hour(sep19['tripduration'])
oct19_avg_dur = mean_hour(oct19['tripduration'])
nov19_avg_dur = mean_hour(nov19['tripduration'])
dec19_avg_dur = mean_hour(dec19['tripduration'])

average_duration_2019 = {'Month': ['January', 'February', 'March', 
                                   'April', 'May', 'June', 
                                   'July', 'August', 'September', 
                                   'October', 'November', 'December'],
                         'Average_Duration': [jan19_avg_dur,
                                              feb19_avg_dur,
                                              mar19_avg_dur,
                                              apr19_avg_dur,
                                              may19_avg_dur,
                                              jun19_avg_dur,
                                              jul19_avg_dur,
                                              aug19_avg_dur,
                                              sep19_avg_dur,
                                              oct19_avg_dur,
                                              nov19_avg_dur,
                                              dec19_avg_dur]}


average_duration_2019 = pd.DataFrame(average_duration_2019)
average_duration_2019.plot.line()


sns.lineplot(x="Month", y="Average_Duration", data=average_duration_2019)
plt.xticks(rotation=30)
plt.title('Typical Trip Duration in 2019')
plt.xlabel('Month')
plt.ylabel('Typical Bike Duration (minutes)')
plt.show()


average_duration_2019
```

```python
jan20_avg_dur = mean_hour(jan20['tripduration'])
feb20_avg_dur = mean_hour(feb20['tripduration'])
mar20_avg_dur = mean_hour(mar20['tripduration'])
apr20_avg_dur = mean_hour(apr20['tripduration'])
may20_avg_dur = mean_hour(may20['tripduration'])
jun20_avg_dur = mean_hour(jun20['tripduration'])
jul20_avg_dur = mean_hour(jul20['tripduration'])
aug20_avg_dur = mean_hour(aug20['tripduration'])
sep20_avg_dur = mean_hour(sep20['tripduration'])
oct20_avg_dur = mean_hour(oct20['tripduration'])
nov20_avg_dur = mean_hour(nov20['tripduration'])
dec20_avg_dur = mean_hour(dec20['tripduration'])

average_duration_2020 = {'Month': ['January', 'February', 'March', 
                                   'April', 'May', 'June', 
                                   'July', 'August', 'September', 
                                   'October', 'November', 'December'],
                         'Average_Duration': [jan20_avg_dur,
                                              feb20_avg_dur,
                                              mar20_avg_dur,
                                              apr20_avg_dur,
                                              may20_avg_dur,
                                              jun20_avg_dur,
                                              jul20_avg_dur,
                                              aug20_avg_dur,
                                              sep20_avg_dur,
                                              oct20_avg_dur,
                                              nov20_avg_dur,
                                              dec20_avg_dur]}
average_duration_2020 = pd.DataFrame(average_duration_2020)
average_duration_2020.plot.line()


sns.lineplot(x="Month", y="Average_Duration", data=average_duration_2020)
plt.xticks(rotation=30)
plt.title('Typical Trip Duration in 2020')
plt.xlabel('Month')
plt.ylabel('Typical Bike Duration (minutes)')
plt.show()


average_duration_2020
```

```python
#Combine all jan19-dec19 dataframes into 1 for 2019
df_2019 = jan19.append(feb19)
df_2019 = df_2019.append(mar19)
df_2019 = df_2019.append(apr19)
df_2019 = df_2019.append(may19)
df_2019 = df_2019.append(jun19)
df_2019 = df_2019.append(jul19)
df_2019 = df_2019.append(aug19)
df_2019 = df_2019.append(sep19)
df_2019 = df_2019.append(oct19)
df_2019 = df_2019.append(nov19)
df_2019 = df_2019.append(dec19)

#Finding total number of starts
grouped_df_start = df_2019.groupby("start station name")
grouped_lists_start = grouped_df_start["bikeid"].apply(list)
grouped_lists_start = grouped_lists_start.reset_index()
grouped_lists_start

#Making df to show numbers of each station start

new_df_start = grouped_lists_start

for i in range(len(grouped_lists_start)):
    new_df_start.loc[i]['bikeid'] = len(grouped_lists_start.loc[i]['bikeid'])

new_df_start.rename(columns = {'bikeid':'start_count', 
                               'start station name':'station'}, inplace = True)
new_df_start

#Finding total number of ends
grouped_df_end = df_2019.groupby("end station name")
grouped_lists_end = grouped_df_end["bikeid"].apply(list)
grouped_lists_end = grouped_lists_end.reset_index()
grouped_lists_end

new_df_end = grouped_lists_end

for i in range(len(grouped_lists_end)):
    new_df_end.loc[i]['bikeid'] = len(grouped_lists_end.loc[i]['bikeid'])

new_df_end.rename(columns = {'bikeid':'end_count', 
                             'end station name':'station'}, inplace = True)


final_df = new_df_start.merge(new_df_end)
cols = ['start_count', 'end_count']
final_df['sum'] = final_df[cols].sum(axis=1)

#Ordering
final_df.sort_values('sum', inplace=True)

final_df['sum'].sum()


final_df.reset_index()

final_df.iloc[48:53].plot(x = 'station',y='sum')

print(final_df.iloc[48:53].values)
```
